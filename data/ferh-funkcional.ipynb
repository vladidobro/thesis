```python
import anal as pm

import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

plt.rc('font', size=8) #controls default text size
plt.rc('axes', titlesize=8) #fontsize of the title
plt.rc('axes', labelsize=8) #fontsize of the x and y labels
plt.rc('xtick', labelsize=8) #fontsize of the x tick labels
plt.rc('ytick', labelsize=8) #fontsize of the y tick labels
plt.rc('legend', fontsize=8) #fontsize of the legend

import locale
locale.setlocale(locale.LC_ALL, 'de_DE.utf8')
plt.rcParams['axes.formatter.use_locale'] = True
```

```python
f = pm.ferh_fm_zeynab()
f.load()
f.preprocess()
f.process()
```

```python
f.analyze()
```

```python
df = pd.DataFrame(f.df['wavelength'])
df['ku'] = f.df['k_canon'].apply(lambda r: r['ku'])
df['phiu'] = f.df['k_canon'].apply(lambda r: r['phiu'])
df['kc'] = f.df['k_canon'].apply(lambda r: r['kc'])
df['phic'] = f.df['k_canon'].apply(lambda r: r['phic'])
df['hext'] = f.df['hext']
df['gamma'] = f.df['sample_rotation']
```

```python
df
```

```python
def zprac(df):
    print(df)
    print(df.kc.mean(),2*df.kc.sem())
    print(df.ku.mean(),2*df.ku.sem())
    print(df.phiu.mean()+180,2*df.phiu.sem())
    print(df.phic.mean()-45,2*df.phic.sem())
```

```python
zprac(df[(df.hext==207) & (df.gamma==0)])
```
